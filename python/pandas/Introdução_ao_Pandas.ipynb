```python
# Importando Bibliotecas
import pandas as pd
```

```python
df = pd.read_csv("Gapminder.csv", error_bad_lines= False, sep=";")
```

```python
# imprimindo 5 primeiras linhas
df.head()
```

```python
# Renomeando colunas de uma DataFrame.
df = df.rename(columns={"country": "Pais",
                   "continent": "continente",
                   "year" : "ano",
                   "lifeExp" : "exp_de_vida",
                   "pop" : "pop_total",
                   "gdpPercap" : "pib"
                   })
```

```python
df.head()
```

```python
# Mostrando as colunas de um dataFrame
df.columns
```

```python
# Mostrando as ultimas linhas
df.tail()
```

```python
# Demostrando informações estátisticas do Dataframe
df.describe()
```

```python
# Imprimindo valores unicos de uma coluna
df["continente"].unique()
```

```python
# Criando uma variável para um continente
oceania = df.loc[df['continente'] == "Oceania"]
```

```python
oceania
```

```python
# Criando um agrupamento e fazendo uma contagem de paises por continentes
df.groupby("continente")['Pais'].nunique()
```

```python
# Criando um agrupamento para calcular a média de idade por ano
df.groupby('ano')["exp_de_vida"].mean()
```
